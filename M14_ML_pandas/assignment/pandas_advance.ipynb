```python
import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({
    'A': [1, None, 3, None, 5],
    'B': ['a', 'b', None, 'd', 'e'],
    'C': [10, 20, 30, 40, 50],
    'D': [100, 200, 300, 400, 500]
})

print("Original DataFrame:")
print(df)
print("\n")

# 1. isna() - Check for missing values
print("1. Using isna() to check missing values:")
print(df.isna())
print("\n")

# 2. fillna() - Fill missing values
print("2. Using fillna() to fill missing values with 0:")
print(df.fillna(0))
print("\n")

# 3. set_index() - Set column as index
print("3. Using set_index() to set column 'C' as index:")
print(df.set_index('C'))
print("\n")

# 4. reset_index() - Reset index to default
df_indexed = df.set_index('C')
print("4. Using reset_index() to reset the index:")
print(df_indexed.reset_index())
print("\n")

# 5. read_csv() - Read a CSV file (creating a small CSV first for demonstration)
import io
csv_data = '''A,B,C
1,x,10
2,y,20
3,z,30'''

# Writing and reading CSV
with open('sample.csv', 'w') as f:
    f.write(csv_data)

print("5. Using read_csv() to read a CSV file:")
print(pd.read_csv('sample.csv'))
print("\n")

# Additional 5 functions:

# 6. describe() - Generate descriptive statistics
print("6. Using describe() to get statistical summary:")
print(df.describe())
print("\n")

# 7. drop() - Drop specified rows or columns
print("7. Using drop() to remove column 'D':")
print(df.drop('D', axis=1))
print("\n")

# 8. groupby() - Group DataFrame by values
df2 = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B', 'A'],
    'Values': [1, 2, 3, 4, 5]
})
print("8. Using groupby() to group and sum values:")
print(df2.groupby('Category')['Values'].sum())
print("\n")

# 9. sort_values() - Sort DataFrame by values
print("9. Using sort_values() to sort by column 'C':")
print(df.sort_values('C', ascending=False))
print("\n")

# 10. value_counts() - Count unique values
df3 = pd.Series(['A', 'B', 'A', 'C', 'B', 'A'])
print("10. Using value_counts() to count occurrences:")
print(df3.value_counts())

```

### Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.

```python
# Create sample data with email addresses
data = {
    'Email': [
        'john.doe@gmail.com',
        'jane.smith@yahoo.com', 
        'bob.wilson@hotmail.com',
        'sarah.jones@outlook.com',
        'mike.brown@company.com',
        'lisa.white@domain.com',
        'david.miller@email.com',
        'emma.davis@example.com',
        'james.taylor@business.com',
        'amy.wilson@website.com'
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Extract username from email addresses
df['username'] = df['Email'].apply(lambda mail: mail.split('@')[0])
df
```

### Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python
# Create sample data with dates and sales
dates = pd.date_range(start='2023-01-01', periods=10, freq='D')
data = {
    'Date': dates,
    'Sales': [120, 135, 95, 110, 145, 125, 160, 140, 130, 150]
}

# Create DataFrame
df = pd.DataFrame(data)
df
```

```python
# Calculate 7-day moving average for Sales column
df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
```

```python
df
```

### Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.


```python
# Create sample DataFrame with dates
date_range = pd.date_range(start='2023-01-01', end='2023-02-15', freq='D')
df = pd.DataFrame({'Date': date_range, 'Value': range(len(date_range))})

# Filter rows between Jan 1 and Jan 31
mask = (df['Date'] >= '2023-01-01') & (df['Date'] <= '2023-01-31')
filtered_df = df[mask]

filtered_df
```
