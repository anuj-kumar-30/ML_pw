# Data Interpolation
It is the process of estimating unknown values within a dataset based on the known values.In python, there are various libraries available that can be used for data interpolation, such numpy, scipy, and pandas.

## 1. Linear Interpolation

```python
import numpy as np
import pandas as pd
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
x,y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
x_new = np.linspace(1,5,10) # create new x values
y_interp = np.interp(x_new, x, y) # interpolate y values
```

```python
plt.scatter(x_new, y_interp)
```

## 2. Cubic interpolation

```python
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
# create a cubic interpolation function
f=interp1d(x,y,kind='cubic')
```

```python
# interpolate the data
x_new = np.linspace(1,5,10)
y_interp = f(x_new)
y_interp
```

```python
x_new
```

```python
plt.scatter(x_new, y_interp)
```

## 3. Polynomial Interpolation

```python
# create some sample data
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
```

```python
# interpolate the data using polynomial interpolation
p = np.polyfit(x, y, 2) # interpolate y values
```

```python
x_new = np.linspace(1,5,15)
y_interp = np.polyval(p, x_new) 
```

```python
x_new, y_interp
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new, y_interp)
```
