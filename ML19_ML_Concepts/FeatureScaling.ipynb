```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
# standardize the total_bill features
scaled_total_bill = scaler.fit_transform(df[['total_bill']])
```

```python
# calculating standardization of total_bill manual using formula
m = df['total_bill'].mean()
s = df['total_bill'].std()
total_bill = list(df['total_bill'])

scaled_bill = []
for i in total_bill:
    z = (i-m)/s
    scaled_bill.append(z)

scaled_bill
```

```python
sns.displot(scaled_bill)
```

```python
sns.displot(df['total_bill'])
```

```python
sns.displot(scaled_total_bill)
```
