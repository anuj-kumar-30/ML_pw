```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,10, 200)
x
```

```python
 
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel('This is my x Data')
plt.ylabel('This is my y Data')
plt.title('Sin Plot of x v/s y')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
plt.scatter(x,y)
```

```python
colors = np.random.rand(50)
sizes = 100*np.random.rand(50)
plt.scatter(x,y, c=colors, s = sizes, alpha=.5)
# alpha: controls the opacity of the data.
```

```python
x = ['a', 'b', 'c', 'd', 'e']
y = np.random.rand(5)
```

```python
plt.bar(x,y) # generally best for categorical vs numerical cols
```

```python
plt.barh(x,y)
```

```python
plt.plot(x,y, '--g')
plt.grid()
plt.figure(figsize=(8,4))
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50)*1000
plt.scatter(x,y , color='blue', s= sizes, alpha=0.5, marker='v')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

# 3-d representation
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```
